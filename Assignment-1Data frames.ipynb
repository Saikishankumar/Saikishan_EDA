```python
name=['Ramesh','Suresh','Satish','kishan','suman','nithin','sandesh','rohith']
age=[30,35,40,30,40,39,36,69,71]
name,age
```

$step-1$

$create$ $dataframe$

```python
pd.DataFrame()
```

$step-2$

$creating$ $dataframes$ $from$ $given$ $list$

```python
pd.DataFrame(zip(name,age))
```

$step-3$

$provide$ $columns$

```python
pd.DataFrame(zip(name,age),columns=['name','age'])
```

$step-4$

$Provide$ $index$

```python
pd.DataFrame(zip(name,age),columns=['name','age'],index=['A','B','C','D','E','F','G','H'])
```

$step-5$

$Add$ $new$ $column$

```python
df=pd.DataFrame(zip(name,age),columns=['name','age'],index=['A','B','C','D','E','F','G','H'])
city_names=["Hyd","Hyd","Hyd","Banglore","USA","Delhi","Mumbai","Pune"]
relations=['not friend','not friend','not friend','friend','school friend','inter friend','inter friend','inter friend']
df["city"]=city_names
df["Relation"]=relations
df
```

$step-6$

$update$ $the$ $exsisting$ $column$

```python
df["name"]=["Ram","Suri","Sati","Kit","Biswas","Goka","Sandy","Rohit"]
df
```

$step-7$

$drop$ $the$ $column$

```python
df.drop("Relation",axis=1,inplace=True)
df
```

$step-8$

$drop$ $the$ $row$

```python
df.drop("H",axis=0,inplace=True)
df
```

$step-9$

$save$ $the$ $dataframe$

```python
df.to_excel("output.xlsx")
```

```python
pd.read_excel("output.xlsx")
```

$step-10$

$remove$ $the$ $index$

```python
df.to_excel("output.xlsx",index=False)
pd.read_excel("output.xlsx")
```
